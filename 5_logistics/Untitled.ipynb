```python
import pandas as pd
from datetime import datetime
```

```python
order_df = pd.read_csv("./data/delivery_orders_march.csv")
order_df.head()
```

```python
order_df["pick"] = pd.to_datetime(order_df["pick"], unit = 's')
order_df["1st_deliver_attempt"] = pd.to_datetime(order_df["1st_deliver_attempt"], unit = 's')
order_df["2nd_deliver_attempt"] = pd.to_datetime(order_df["2nd_deliver_attempt"], unit = 's')
order_df.head()
```

```python
order_df["pick"] = order_df["pick"].apply(lambda x: x.replace(hour = 0, minute= 0, second = 0))
order_df["1st_deliver_attempt"] = order_df["1st_deliver_attempt"].apply(lambda x: x.replace(hour = 0, minute= 0, second = 0))
order_df["2nd_deliver_attempt"] = order_df["2nd_deliver_attempt"].apply(lambda x: x.replace(hour = 0, minute= 0, second = 0))
order_df.head()
```

```python
sla = {
    'metro manila': {
        'metro manila': 3,
        'luzon': 5,
        'visayas': 7,
        'mindanao': 7
    },
    'luzon': {
        'metro manila': 5,
        'luzon': 5,
        'visayas': 7,
        'mindanao': 7
    },
    'visayas': {
        'metro manila': 7,
        'luzon': 7,
        'visayas': 7,
        'mindanao': 7
    },
    'mindanao': {
        'metro manila': 7,
        'luzon': 7,
        'visayas': 7,
        'mindanao': 7
    }
}
locations = list(sla.keys())
pub_hols = [datetime(2020, 3, 25), datetime(2020, 3, 30), datetime(2020, 3, 31)]

print(sla)
print(locations)
print(pub_hols)
```

```python
def get_location(add):
    add = add.lower()
    for idx, loc in enumerate(locations):
        add = add.replace(loc, "<<{}".format(idx))
    
    tokens = add.split()
    for token in tokens:
        try:
            if token[:2] == "<<":
                loc_idx = int(token[-1])
                return locations[loc_idx]
        except Exception:
            pass
    return None

order_df['origin'] = order_df['selleraddress'].apply(get_location)
order_df['dest'] = order_df['buyeraddress'].apply(get_location)
order_df.head()
```

```python
def num_days_between(start, end):
    num_weeks, remainder = divmod((end-start).days, 7)
    if (6 - start.weekday() ) % 7 <= remainder:
        return num_weeks + 1
    else:
        return num_weeks
    
def nump_1(row):
    num_sundays = num_days_between(row['pick'], row['1st_deliver_attempt'])
    
    after_start = [dt >= row['pick'] for dt in pub_hols]
    before_end = [dt <= row['1st_deliver_attempt'] for dt in pub_hols]
    num_pub_hols = sum([x[0] and x[1] for x in zip(after_start, before_end)])

    return row['1st_deliver_attempt'] - row['pick'] - pd.Timedelta(days = num_sundays + num_pub_hols)

order_df['p-1'] = order_df.apply(lambda x: nump_1(x), axis = 1)
order_df.head()
```

```python
def num1_2(row):
    if row['2nd_deliver_attempt'].isna():
        return None

    num_sundays = num_days_between(row['1st_deliver_attempt'], row['2nd_deliver_attempt'])
    
    after_start = [dt >= row['1st_deliver_attempt'] for dt in pub_hols]
    before_end = [dt <= row['2nd_deliver_attempt'] for dt in pub_hols]
    num_pub_hols = sum([x[0] and x[1] for x in zip(after_start, before_end)])

    return row['2nd_deliver_attempt'] - row['1st_deliver_attempt'] - pd.Timedelta(days = num_sundays + num_pub_hols)

order_df["1_2"] = order_df.apply(lambda x: num1_2(x), axis = 1)
order_df.head()
```

```python
order_df.head()
```

```python
output = []
for idx, row in order_df.iterrows():
    p_1_limit = sla[row['origin']][row['dest']]
    if row['p-1'].days > p_1_limit:
        output.append([row['orderid'], 1])
    elif pd.notnull(row["1_2"]) and row['1_2'].days > 3:
        output.append([row['orderid'], 1])
    else:
        output.append([row['orderid'], 0])
        
output_df = pd.DataFrame(output, columns = ['orderid', 'is_late'])
output_df.head()
```

```python
output_df.to_csv("./output.csv", index = False)
```
